```python
import requests

# RapidAPI parameters
RAPIDAPI_HOST = 'api-football-v1.p.rapidapi.com'
# RAPIDAPI_KEY = '8cce961234msh8550452e950ada9p179b4bjsn57eb3dbb9c26'
RAPIDAPI_KEY = os.getenv('API_KEY')
# API endpoints for each league
endpoints = {
    'La Liga': 'https://api-football-v1.p.rapidapi.com/v3/teams?league=140&season=2024',
    # 'Premier League': 'https://api-football-v1.p.rapidapi.com/v3/leagues',
    # 'Bundesliga': 'https://api-football-v1.p.rapidapi.com/v3/teams?league=BL1&seasons=2020,2021,2022,2023,2024',
    # 'Serie A': 'https://api-football-v1.p.rapidapi.com/v3/teams?league=SA&seasons=2020,2021,2022,2023,2024'
}

headers = {
    'x-rapidapi-host': RAPIDAPI_HOST,
    'x-rapidapi-key': RAPIDAPI_KEY
}

def fetch_teams():
    all_teams = []

    # Iterate through each league and make API requests
    for league, endpoint in endpoints.items():
        response = requests.get(endpoint, headers=headers)
        if response.status_code == 200:
            print(response.json())
            
        else:
            print(f"Failed to fetch data for {league}: {response.status_code}")

    return all_teams

if __name__ == '__main__':
    teams = fetch_teams()
    print(f"Total teams fetched: {len(teams)}")
    # Process teams further (store in database, etc.)
```
